```julia
mn = [20, 35, 30, 35, 27,25, 32, 34, 20, 25]
sx = repeat(["Men", "Women"], inner = 5)
std = [2, 3, 4, 1, 2, 3, 5, 2, 3, 3]
nam = repeat("G" .* string.(1:5), outer = 2)

using StatPlots
groupedbar(nam, mn, yerr = std, group = sx, ylabel = "Scores", 
        title = "Scores by group and gender")
```
